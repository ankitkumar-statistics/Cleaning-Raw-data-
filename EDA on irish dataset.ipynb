### Iris Dataset
* It contains five columns
* Petal Length, Petal Width, Sepal Length, Sepal Width and Species Type
* Iris is a flowering plant

#### Checking Duplicates

```python
df.drop_duplicates(subset = "Species")
# here we are removing all the duplicates value based on Species 
# so here we get only 3 sample data because ther are only three unique species 
```

```python
df.value_counts('Species')
# here all species contain equal amount of roes
```

#### Data visualization
* Here our target variable will be 'Species'
* Here we will use matplotlib & seaborn library 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x = "Species", data = df)
plt.show()
```

#### Compering Sepal Length and Speal Width

```python
sns.scatterplot(x = "SepalLengthCm", y="SepalWidthCm",
               hue = "Species", data = df)

plt.legend(bbox_to_anchor = (1, 1), loc = 2)
plt.show()
```

* here each point is representing flower with its Sepal Length and Sepal Width
* Colour of the point tells us the Species of the flower
* from above visuals we can say that Species setosa has smaller Sepal Length but large Sepal widths
* Versicolor mostly between these Species
* Virginica has large Saple Length but Small Sepal width

#### Compering Petal length and Petal Width

```python
sns.scatterplot(x = "PetalLengthCm", y = "PetalWidthCm",
               hue = "Species", data = df)
plt.legend(bbox_to_anchor = (1, 1), loc = 2)
plt.show()
```

* from above visuals we can say that species setosa has smaller petal length and width
* versicolor peatl length and width lies in the middle of other two species
* virginica has largest petal length and width

```python
sns.pairplot(df.drop(['Id'], axis=1), hue="Species", height=2)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (10, 10))
axes[0,0].set_title("Sepal Length")
axes[0,0].hist(df["SepalLengthCm"], bins = 7)

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(df["SepalWidthCm"], bins = 5)

axes[1,0].set_title("Petal Length")
axes[1,0].hist(df["PetalLengthCm"], bins = 6)

axes[1,1].set_title("Petal Width")
axes[1,1].hist(df["PetalWidthCm"], bins = 6)

plt.show()
```

```python
plot = sns.FacetGrid(df ,hue = "Species")
plot.map(sns.histplot, "SepalLengthCm").add_legend()

plot = sns.FacetGrid(df, hue = "Species")
plot.map(sns.histplot, "SepalWidthCm").add_legend()

plot = sns.FacetGrid(df, hue = "Species")
plot.map(sns.histplot, "PetalLengthCm").add_legend()

plot = sns.FacetGrid(df, hue = "Species")
plot.map(sns.histplot, "PetalWidthCm").add_legend()

plt.show()
```

```python
columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for column in columns:
    plot = sns.FacetGrid(df, hue = "Species")
    plot.map(sns.histplot, column, kde = True).add_legend()

plt.show()
```

```python
correlation_matrix = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr(method="pearson")
correlation_matrix
```

```python
def graph(y):
    sns.boxplot(x = "Species", y = y, data =df)

plt.figure(figsize = (10, 10))

plt.subplot(221)
graph('SepalLengthCm')

plt.subplot(222)
graph('SepalWidthCm')

plt.subplot(223)
graph('PetalLengthCm')

plt.subplot(224)
graph('PetalWidthCm')
```
